```python
df1 = pd.read_csv("/content/drive/MyDrive/linha1.csv", decimal=',')
df1 = df1.astype(float)
```

```python
df2 = pd.read_csv("/content/drive/MyDrive/linha2.csv", decimal=',')
df2 = df2.astype(float)
```

```python
df3 = pd.read_csv("/content/drive/MyDrive/linha3.csv", decimal=',')
df3 = df3.astype(float)
```

```python
df4 = pd.read_csv("/content/drive/MyDrive/linha4.csv", decimal=',')
df4 = df4.astype(float)
```

```python

#LINHA3
fig, g1 = plt.subplots()
g2=g1.twinx()
sns.lineplot(data=df3, y=' Velocity [ m s^-1 ]', x='X [ m ]', ax=g1, color='black')
sns.lineplot(data=df3, y=' Pressure [ Pa ]', x='X [ m ]', ax=g2, color='red')
g1.set_ylabel(" Velocity [ m s^-1 ]", color = 'black')
g1.set_xlabel("X [ m ]")
g1.set_xticklabels(df3['X [ m ]'], rotation='vertical', size=9)
g2.set_ylabel(' Pressure [ Pa ]', color='red')
plt.grid()
plt.show()

fig, g1 = plt.subplots()
g2=g1.twinx()
sns.lineplot(data=df3, y=' Velocity [ m s^-1 ]', x=' Y [ m ]', ax=g1, color='black')
sns.lineplot(data=df3, y=' Pressure [ Pa ]', x=' Y [ m ]', ax=g2, color='red')
g1.set_ylabel(" Velocity [ m s^-1 ]", color = 'black')
g1.set_xlabel(" Y [ m ]")
g1.set_xticklabels(df3[' Y [ m ]'], rotation='vertical', size=9)
g2.set_ylabel(' Pressure [ Pa ]', color='red')
plt.grid()
plt.show()

fig, g1 = plt.subplots()
g2=g1.twinx()
sns.lineplot(data=df3, y=' Velocity [ m s^-1 ]', x=' Z [ m ]', ax=g1, color='black')
sns.lineplot(data=df3, y=' Pressure [ Pa ]', x=' Z [ m ]', ax=g2, color='red')
g1.set_ylabel(" Velocity [ m s^-1 ]", color = 'black')
g1.set_xlabel(" Z [ m ]")
g1.set_xticklabels(df3[' Z [ m ]'], rotation='vertical', size=9)
g2.set_ylabel(' Pressure [ Pa ]', color='red')
plt.grid()
plt.show()
```
